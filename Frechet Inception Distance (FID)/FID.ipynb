# Plotting Values

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    'Class':['Covid', 'Viral Pneumonia', 'Covid', 'Viral Pneumonia', 'Covid', 'Viral Pneumonia'],
    'GAN':['DCGAN', 'DCGAN', 'WGAN', 'WGAN', 'WGAN-GP', 'WGAN-GP'],
    'FID':[83.65, 198.16, 55.50, 90.45, 70.21, 103.86]
    })
```

```python
df
```

```python
fig, axes = plt.subplots(1, 1, figsize=(20, 12))

ax1 = plt.subplot(1,1, 1)

plot_1 = sns.barplot(x = 'Class', y = 'FID', hue = 'GAN',data = df\
                     , capsize=.2, ax = ax1)

for container in ax1.containers:
    ax1.bar_label(container, fontsize = 25)


ax1.tick_params(axis ='x', labelbottom=True, labelsize = 40, pad = 0) 
ax1.tick_params(axis ='y', labelbottom=True, labelsize = 40)
ax1.set_xlabel('')
ax1.set_ylabel('')
plt.legend(fontsize = 25, loc = 'upper left')
ax1.set_ylim([0, 220])


# set various colors
ax1.spines['bottom'].set_color('black')
ax1.spines['top'].set_color('black') 
ax1.spines['right'].set_color('black')
ax1.spines['right'].set_linewidth(2)
ax1.spines['top'].set_linewidth(2)
ax1.spines['bottom'].set_linewidth(2)
ax1.spines['left'].set_color('black')
ax1.spines['left'].set_lw(2)
ax1.yaxis.set_major_locator(plt.MaxNLocator(8))

# for ax1 
ax1.tick_params(which='both', width=5)
ax1.tick_params(which='major', length=30, axis = 'x')
ax1.tick_params(which='major', length=20, axis = 'y')
ax1.tick_params(which='minor', length=15)
ax1.grid(True, which = 'major', alpha = 0.2, linestyle='--', linewidth = 1, color = 'black')

fig.text(0.025, 0.5, 'Frechet Inception Distance (FID)', ha='center', va='center', rotation='vertical', fontsize = 40, color = 'black')

plt.savefig('FID_GANS.jpg',  bbox_inches='tight', dpi = 500)
```
